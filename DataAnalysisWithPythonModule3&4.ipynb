## Module 4 Project 2 ~ Model Development 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore", category=UserWarning) 
%matplotlib inline

import sys
sys.path.append(r'C:\\Code\\Python\\HelperFunctions')
from helperFunctions import download_file

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
file_name ="laptops.csv"
download_file(file_path, file_name)
```

```python
df = pd.read_csv('laptops.csv', header=0)
df.head()
```

```python
cpu_freq_model = LinearRegression()
X_cpu_freq = df[['CPU_frequency']]
Y_cpu_freq = df['Price']
cpu_freq_model.fit(X_cpu_freq, Y_cpu_freq)

```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
cpu_yhat = cpu_freq_model.predict(X_cpu_freq)
cpu_mse = mean_squared_error(Y_cpu_freq, cpu_yhat)
cpu_r2 = r2_score(Y_cpu_freq, cpu_yhat)
print("The mean square error of CPU frequency and Price is: ", cpu_mse)
print("The R2 score of CPU frequency and Price is: ", cpu_r2)
```

```python
import seaborn as sns
ax1 = sns.distplot(Y_cpu_freq, hist=False, color='blue', label='Actual Values')
sns.distplot(cpu_yhat, hist=False, color='red', label='Fitted Values' , ax=ax1)
plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price')
plt.ylabel('Proportion of laptops')
plt.legend(['Actual Value', 'Predicted Value'])
plt.show()
```

```python
multi_laptop_model = LinearRegression()
z_mutli = df[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']]
y_multi = df['Price']
multi_laptop_model.fit(z_mutli, y_multi)
multi_yhat = multi_laptop_model.predict(z_mutli)
multi_mse = mean_squared_error(y_multi, multi_yhat)
multi_r2 = r2_score(y_multi, multi_yhat)
print("The mean square error of multi laptop model is: ", multi_mse)
print("The R2 score of multi laptop model is: ", multi_r2)
```

```python
import seaborn as sns
ax1 = sns.distplot(y_multi, hist=False, color='blue', label='Actual Values')
sns.distplot(multi_yhat, hist=False, color='red', label='Fitted Values' , ax=ax1)
plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price')
plt.ylabel('Proportion of laptops')
plt.legend(['Actual Value', 'Predicted Value'])
plt.show()
```

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
```

```python
x_poly = df['CPU_frequency']
y_poly = df['Price']
poly_2 = np.polyfit(x_poly, y_poly, 2)
poly_3 = np.polyfit(x_poly, y_poly, 5)
poly_4 = np.polyfit(x_poly, y_poly, 6)
PlotPolly(np.poly1d(poly_2), x_poly, y_poly, 'CPU_frequency')
PlotPolly(np.poly1d(poly_3), x_poly, y_poly, 'CPU_frequency')
PlotPolly(np.poly1d(poly_4), x_poly, y_poly, 'CPU_frequency')
```

```python
yhat_poly_2 = np.poly1d(poly_2)(x_poly)
yhat_poly_3 = np.poly1d(poly_3)(x_poly)
yhat_poly_4 = np.poly1d(poly_4)(x_poly)


poly_2_mse = mean_squared_error(y_poly, yhat_poly_2)
poly_2_r2 = r2_score(y_poly, yhat_poly_2)
print("The mean square error of poly 2 degree laptop model is: ", poly_2_mse)
print("The R2 score of poly 2 degree laptop model is: ", poly_2_r2)
poly_3_mse = mean_squared_error(y_poly, yhat_poly_3)
poly_3_r2 = r2_score(y_poly, yhat_poly_3)
print("The mean square error of poly 3 degree laptop model is: ", poly_3_mse)
print("The R2 score of poly 3 degree laptop model is: ", poly_3_r2)
poly_4_mse = mean_squared_error(y_poly, yhat_poly_4)
poly_4_r2 = r2_score(y_poly, yhat_poly_4)
print("The mean square error of poly 4 degree laptop model is: ", poly_4_mse)
print("The R2 score of poly 4 degree laptop model is: ", poly_4_r2)
```

```python
pipe = Pipeline([('scale', StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())])
pipe.fit(z_mutli, y_multi)
y_pipe = pipe.predict(z_mutli)
```

```python
pipe_mse = mean_squared_error(y_poly, y_pipe)
pipe_r2 = r2_score(y_poly, y_pipe)
print("The mean square error of the pipeline model is: ", pipe_mse)
print("The R2 score of the pipeline model is: ", pipe_r2)
```

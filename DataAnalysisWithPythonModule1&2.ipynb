## Module 2 Project 1 ~ Data Wrangling

```python
import pandas as pd
import matplotlib.pylab as plt
import sys

sys.path.append(r'C:\\Code\\Python\\HelperFunctions')
from helperFunctions import download_file

file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
file_name="usedcars.csv"

download_file(url=file_path, filename=file_name)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv('usedcars.csv', names=headers)
```

```python
df.head(5)
```

### Data Grooming

```python
import numpy as np
## Replace ? to N
df.replace("?", np.nan, inplace=True)
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

```python
## Update Certain Rows with the Mean of the Column
# for numerical data that isn't overly significant to the model
columns = [
    'normalized-losses',
    'stroke',
    'bore',
    'horsepower',
    'peak-rpm'
]

for column in columns:
    try: 
        avg_column = df[column].astype("float").mean(axis=0)
        df[column].replace(np.nan, avg_column, inplace=True)
    except ValueError as e:
        continue
```

```python
updated_missing_data = df.isnull() 
for column in updated_missing_data.columns.values.tolist():
    print (updated_missing_data[column].value_counts())
    print("")
```

```python
highest_freq = df['num-of-doors'].value_counts().idxmax()
print(highest_freq)
```

```python
## Replace by highest frequency
# for qualitative data with significant bias
columns = [
    'num-of-doors'
]

for column in columns:
    try: 
        highest_freq = df[column].value_counts().idxmax()
        df[column].replace(np.nan, highest_freq, inplace=True)
    except ValueError as e:
        continue
```

```python
## Drop rows with missing price data
# do this since price is the predictive output so speculating may affect model
columns = [
    'price'
]

for column in columns:
    try: 
        df.dropna(subset=[column], axis=0, inplace=True)
        df.reset_index(drop=True, inplace=True)
    except ValueError as e:
        continue
```

```python
df.head()
```

```python
df.dtypes
```

```python
# type conversion
float_list = ['bore', 'stroke', 'price', 'peak-rpm']
int_list = ['normalized-losses']
for column in float_list:
    try:
        df[column] = df[column].astype('float')
    except ValueError as e:
        print(e)

for column in int_list:
    try:
        df[column] = df[column].astype('int')
    except ValueError as e:
        print(e)
```

```python
df.dtypes
```

```python
# add city liter per 100 km
df['city-L/100km'] = 235/df['city-mpg']
df['highway-L/100km'] = 235/df['highway-mpg']
df.head()
```

```python
## Data Normalization
for column in ['length', 'width', 'height']:
    df[column] = df[column]/df[column].max()
```

```python
## Bining
df['horsepower'] = df['horsepower'].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(df['horsepower'].min(),df['horsepower'].max(), 4)
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
```

```python
df.head(20)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
## Using Dummies
# use dummies to split categorical data into numerical data for modeling
dummy_variable1 = pd.get_dummies(df['fuel-type'], dtype=int)
dummy_variable1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable1.head()
```

```python
df = pd.concat([df, dummy_variable1], axis=1)
df.drop('fuel-type', axis=1, inplace=True)
```

```python
df.head()
```

```python
dummy_variable2 = pd.get_dummies(df['aspiration'], dtype=int)
dummy_variable2.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
df = pd.concat([df, dummy_variable2], axis=1)
df.drop('aspiration', axis=1, inplace=True)
```

```python
df.to_csv('groomed_automobiles.csv')
```

## Module 2 - Project 2 - Data Wrangling 2

```python
df = pd.read_csv('laptops.csv', header=0, index_col=0)
df.head()
df.info()
```

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts())
    print('')
```

```python
avg_column = df['Weight_kg'].astype('float').mean()
df['Weight_kg'].replace(np.nan, avg_column, inplace=True)
df.head()
```

```python
most_freq = df['Screen_Size_cm'].value_counts().idxmax()
df['Screen_Size_cm'].replace(np.nan, most_freq, inplace=True)
df.head()
```

```python
df['Weight_kg'] = df['Weight_kg'].astype('float')
df['Screen_Size_cm'] = df['Screen_Size_cm'].astype('float')
```

```python
df['Weight_lbs'] = df['Weight_kg'].astype('float')*2.205
df['Screen_Size_inch'] = df['Screen_Size_cm'].astype('float')/2.54
```

```python
# Normalize CPU_Frequency
df['CPU_frequency'] = df['CPU_frequency'] / df['CPU_frequency'].max()
```

```python
#Binning
bins = np.linspace(df['Price'].min(), df['Price'].max(), 4)
labels = ['Low', 'Medium', 'High']
df['Price-binned'] = pd.cut(df['Price'], bins, labels=labels, include_lowest=True)
```

```python
df.head()
```

```python
df[['Screen_Size_inch']] = np.round(df[['Screen_Size_inch']],2)
df[['Weight_lbs']] = np.round(df[['Weight_lbs']],2)
```

```python
# Indicator variables
for column in ['Screen']:
    dummies = pd.get_dummies(df[column], dtype=int)
    rename_dict = {x : str.replace(column + '-' + x, ' ','_') for x in dummies.columns.values.tolist()}
    print(rename_dict)
    dummies.rename(columns=rename_dict, inplace=True)
    df = pd.concat([df, dummies], axis=1)
    df.drop(column, axis=1, inplace=True)

df.head()
```
